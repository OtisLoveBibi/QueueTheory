```python
#將所有檔案跟Sheet合併成一個

import glob
import pandas as pd
path = r"Supplementary DATA on Queues"
data = []
for excel_file in glob.glob(f'{path}/[!~]*.xlsx'):
    dfs = pd.read_excel(excel_file, sheet_name=None).values()
    data.extend(dfs)
df = pd.concat(data, ignore_index=True)
df.to_excel("allData.xlsx", index=False)
```
